```python
import matplotlib.pyplot as plt

missing = df.isnull().sum()

if (missing > 0).sum() == 0:
    print("✅ No missing values found in the dataset.")
else:
    missing[missing > 0].plot(kind="bar")
    plt.title("Missing Values Count")
    plt.ylabel("Count")
    plt.show()
```

```python
categorical_cols = [
    "Ship Mode", "Segment", "Country", "City", "State",
    "Region", "Category", "Sub-Category", "Payment Mode", "Returns"
]

for col in categorical_cols:
    df[col] = df[col].astype("category")
```

```python
df.memory_usage(deep=True)
```

```python
df["Returns"].value_counts()
```

```python
df["Returns"] = df["Returns"].map({"Yes": 1, "No": 0})
```

```python
df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month
df["Day"] = df["Order Date"].dt.day
df["Weekday"] = df["Order Date"].dt.day_name()
```

```python
df["Shipping Days"] = (df["Ship Date"] - df["Order Date"]).dt.days
```

```python
# Negative profit (important business insight)
df[df["Profit"] < 0].head()

# Zero or negative sales (should not happen)
df[df["Sales"] <= 0]

# Shipping issues
df[df["Shipping Days"] < 0]
```

```python
text_cols = ["City", "State", "Product Name"]

for col in text_cols:
    df[col] = df[col].str.strip()
```

```python
for col in ["Category", "Sub-Category", "Region", "Segment"]:
    print(f"\n{col}:")
    print(df[col].unique())
```

```python
df[["Sales", "Profit", "Quantity"]].describe()
```

```python
import matplotlib.pyplot as plt

df["Sales"].plot(kind="box")
plt.title("Sales Outliers")
plt.show()
```

```python
# Profit should logically relate to Sales
df[df["Profit"] > df["Sales"]].head()

# Shipping days sanity
df[df["Shipping Days"] > 30]
```

```python
df["Profit Margin"] = df["Profit"] / df["Sales"]
```

```python
df["Avg Order Value"] = df["Sales"] / df["Quantity"]
```

```python
customer_orders = df.groupby("Customer ID")["Order ID"].nunique()
```

```python
customer_sales = df.groupby("Customer ID")["Sales"].sum()
```

```python
region_profit = df.groupby("Region")["Profit"].sum()
region_sales = df.groupby("Region")["Sales"].sum()
```

```python
df["Month Name"] = df["Order Date"].dt.month_name()
df["Quarter"] = df["Order Date"].dt.quarter
```
